```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
data.info
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
data.drop('Review Lowercase', axis=1, inplace=True)
```

```python
data.head()
```

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove('not')
```

```python
data['review_no_stopwords'] = data['review_lowercase'].apply(lambda x : ' '.join([word for word in x.split() if word not in en_stopwords]))
```

```python
data['review_no_stopwords'].head()
```

```python
data['review_no_stopwords'][0]
```

```python
data['review_no_stop_no_punc'] = data.apply(lambda x : re.sub(r"[*]", "star", x["review_no_stopwords"]), axis=1)
```

```python
data.head()
```

```python
data['review_no_stop_no_punc'] = data.apply(lambda x : re.sub(r"[^\w\s]", "", x["review_no_stop_no_punc"]), axis=1)
```

```python
data['review_no_stop_no_punc'][0]
```

```python
data['tokenized'] = data.apply(lambda x : word_tokenize(x['review_no_stop_no_punc']), axis=1)
```

```python
data['tokenized'][0]
```

```python
ps = PorterStemmer()
```

```python
data['stemmed'] = data['tokenized'].apply(lambda x : [ps.stem(word) for word in x])
```

```python
data['stemmed'][0]
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
data['lemmatized'] = data['tokenized'].apply(lambda x : [lemmatizer.lemmatize(word) for word in x])
```

```python
data.lemmatized[0]
```

```python
tokens_clean = sum(data['lemmatized'], [])
```

```python
unigrams = pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts()
```

```python
unigrams
```

```python
bigrams =  pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts()
```
