```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

<h3>Load Data</h3>

```python
bbc_data = pd.read_csv("bbc_news.csv")
```

```python
bbc_data.head()
```

```python
bbc_data.info()
```

```python
titles = pd.DataFrame(bbc_data['title'])
titles.head()
```

<h3>Clean Data</h3>

```python
#lowercase-> stopwords remove  -> punctuation remove -> tokenize -> lemmatize
```

```python
#lowercase
titles['lowercase'] = titles['title'].str.lower()
titles.head()
```

```python
#stopwords remove
en_stopwords = stopwords.words("english")
titles['no_stopwords'] = titles['lowercase'].apply(lambda x : ' '.join([word for word in x.split() if word not in en_stopwords]))
titles.head()
```

```python
#punc remove
titles['no_stop_punc'] = titles['no_stopwords'].apply(lambda x : re.sub(r"([^\w\s])","", x))
titles['no_stop_punc']
```

```python
#tokenize
titles['tokenize_clean'] = titles.apply(lambda x : word_tokenize(x['no_stop_punc']), axis=1)
titles['tokenize_raw'] = titles.apply(lambda x : word_tokenize(x['title']), axis=1)
titles['tokenize_clean']
```

```python
#lemmatizer
lematizer = WordNetLemmatizer()
titles['token_clean_lemmatized'] = titles['tokenize_clean'].apply(lambda tokens : [lematizer.lemmatize(token) for token in tokens])
titles.head()
```

```python
tokens_raw_list = sum(titles['tokenize_raw'],[])
tokens_clean_list = sum(titles['token_clean_lemmatized'],[])
```

<h3>Pos Tagging</h3>

```python
nlp = spacy.load("en_core_web_sm")
```

```python
spacy_doc = nlp(' '.join(tokens_raw_list))
```

```python
pos_df = pd.DataFrame(columns=['token', 'pos_tag'])
for token in spacy_doc:
    pos_df = pd.concat([pos_df, pd.DataFrame.from_records([{'token': token.text, 'pos_tag': token.pos_}])], ignore_index=True)
```

```python
pos_df_count = pos_df.groupby(['token', 'pos_tag']).size().reset_index(name='counts').sort_values(by="counts", ascending=False)
pos_df_count
```

```python
nouns = pos_df_count[pos_df_count.pos_tag == "NOUN"][:10]
nouns
```

```python
verbs = pos_df_count[pos_df_count.pos_tag == "VERB"][:10]
verbs
```

<h3>NER</h3>

```python
ner_df = pd.DataFrame(columns=['token','ner_tag'])
```

```python
for token in spacy_doc.ents:
    if pd.isna(token.label_) is False:
        ner_df = pd.concat([ner_df, pd.DataFrame.from_records([{'token': token.text, 'ner_tag': token.label_}])], ignore_index=True)
ner_df
```

```python
ner_df_count = ner_df.groupby(['token', 'ner_tag']).size().reset_index(name="counts")
ner_df_count
```
